```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('MainData.csv', encoding='cp949')
df = df.dropna()
df
```

```python
#x = df[['층','임대면적','보증금','건축년도','경제활동인구','기준금리','정책성향','아파트매매실거래지수','회사채수익률','소비자물가지수','통화량','광공업지수']]
x = df['보증금'].values
x = x.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
x = sc.fit_transform(x)
```

```python
x = pd.DataFrame(data=x)
#x.columns = ['층','임대면적','보증금','건축년도','경제활동인구','기준금리','정책성향','아파트매매실거래지수','회사채수익률','소비자물가지수','통화량','광공업지수']
```

```python
x
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot  as plt
import seaborn as sns

# create model and prediction
model = DBSCAN(eps=0.5,min_samples=3)
predict = pd.DataFrame(model.fit_predict(x))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([x,predict],axis=1)
```

```python
r.to_csv('DBSCAN.csv',encoding='cp949')
```
